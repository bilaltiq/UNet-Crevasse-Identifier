```python
#########################################
### Opening net cdf file to determine shape and structure
##########################################
f_plastic_runs = []

for filename in os.listdir(fine_mesh):
    if filename.startswith('f-m8'):
        file_path = os.path.join(fine_mesh, filename)
        file_path = xr.open_dataset(file_path)
        f_plastic_runs.append(file_path)
```

```python
f_plastic_runs[0]
x_ndh = f_plastic_runs[0]['x']
y_ndh = f_plastic_runs[0]['y']
```

```python
# x_ndh and y_ndh are in polar stereographic
## switch to coordinate system of ASF search to download in region
# Step 1: Compute the bounding box of your model area
xmin = x_ndh.min()
xmax = x_ndh.max()
ymin = y_ndh.min()
ymax = y_ndh.max()

transformer = Transformer.from_crs("EPSG:3031", "EPSG:4326", always_xy=True)



lon_min, lat_min = transformer.transform(xmin, ymin)
lon_max, lat_max = transformer.transform(xmax, ymax)
```

```python
# Function to check if a .tiff file already exists
def check_if_file_exists(filename, data_dir):
    file_path = os.path.join(data_dir, filename)
    return os.path.isfile(file_path)
```

```python
# Function to check if a zip file has already been unzipped
def check_if_unzipped(zip_filename, data_dir):
    # Construct the expected unzipped folder name (e.g., from the zip file name)
    safe_folder = os.path.join(data_dir, zip_filename.replace(".zip", ".SAFE"))
    return os.path.isdir(safe_folder)
```

```python
if download_data_flag == 1:
    session = asf.ASFSession()

    # Loop through the filtered results
    for result in filtered_results:
        filename = f"{result.properties['fileID']}.tiff"

        # Check if the file already exists
        if check_if_file_exists(filename, data_dir):
            print(f"File {filename} already exists. Skipping download.")
        else:
            print(f"Downloading {filename}...")
            # This uses NASA Earth Data Login to authenticate and download
            result.download(path=data_dir, session=session)
```

```python
if unzip_data_flag == 1:
    zip_files = sorted(glob.glob(data_dir+'/*.zip'))

    for zf in zip_files:
    # Check if the folder has already been unzipped
        if check_if_unzipped(zf, data_dir):
            print(f"File {zf} already exists. Skipping download.")
        else:
            print(f"Unzipping {zf}...")
            try:
                zip_directory = '/'.join(zf.split('/')[:-1])

                subprocess.run(['unzip', zf, '-d', zip_directory], check=True)
                print(f"Unzipped {zf}")
            except subprocess.CalledProcessError as e:
                    print(f"Error unzipping {zf}: {e}")
```
